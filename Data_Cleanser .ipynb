# Handling Missing Values

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
import numpy as np
from scipy.stats import zscore
from scipy.stats.mstats import winsorize
```

```python
#Importing dataset
df = pd.read_csv("patient_health_records.csv")
df
```

```python
print("\nMissing Values Count:")
print(df.isnull().sum())
print("\nPercentage of Missing Values:")
print((df.isnull().sum() / len(df) * 100).round(2))
```

## Simple Imputer(Numerical) 

```python
# using simple imputer numrical for filling bmi missing valus using mean
num_imputer = SimpleImputer(strategy='mean')
df[['bmi']] = num_imputer.fit_transform(df[['bmi']])
print("After BMI Mean Imputation:")
df
```

## Simple Imputer(Categorical)

```python
# using simple imputer categorical for filling bmi missing valus using mean
cat_imputer = SimpleImputer(strategy='most_frequent')
df[['region']] = cat_imputer.fit_transform(df[['region']])
print("After Region Most Frequent Imputation:")
df
```

## Most Frequent Imputation

```python
df['gender'] = df['gender'].fillna(df['gender'].mode()[0])
df
```

## Missing Indicator + Random Sampling

```python
# Create indicator column
df['age_missing'] = df['age'].isnull().astype(int)

# Random sampling imputation
age_non_null = df['age'].dropna()

df.loc[df['age'].isnull(), 'age'] = np.random.choice(age_non_null, size=df['age'].isnull().sum())
df
```

## KNN Imputer (Multivariate)

```python
#Using knn to fill age, bmi, blood pressure, cholestrol, glucose columns
knn_imputer = KNNImputer(n_neighbors=5)
num_cols = ['age', 'bmi', 'blood_pressure', 'cholesterol', 'glucose']
df[num_cols] = knn_imputer.fit_transform(df[num_cols])
df
```

## MICE Algorithm

```python

mice_imputer = IterativeImputer(random_state=42)
df[num_cols] = mice_imputer.fit_transform(df[num_cols])
df
```

# Handling Outliers

```python
df_outliers = df.copy()
```

## Z-Score Method

```python


# Numerical columns
num_cols = ['bmi', 'blood_pressure', 'cholesterol', 'glucose']

# Compute Z-scores
z_scores = np.abs(zscore(df_outliers[num_cols], nan_policy='omit'))

# Convert to DataFrame
z_df = pd.DataFrame(z_scores, columns=num_cols)

# Identify outliers (if any column > 3)
outlier_mask = (z_df > 3).any(axis=1)

# Add columns to original DataFrame
df_outliers['z_score_flag'] = outlier_mask

print("Using scipy zscore():")
print(f"  Outliers found : {outlier_mask.sum()}")
print(f"  Original rows  : {len(df_outliers)}")

# Show outliers
print("\nOutlier rows:")
print(df_outliers[df_outliers['z_score_flag']].head())

# Remove outliers
df_zscore = df_outliers[~df_outliers['z_score_flag']].copy()

print(f"\nDataFrame cleaned: {len(df_outliers)} → {len(df_zscore)} rows")
```

## IQR Method

```python
df_iqr = df_outliers.copy()  # Create a copy so original data remains unchanged

# Loop through each numerical column to detect and remove outliers
for col in num_cols:
    
    # Calculate first quartile (25th percentile)
    Q1 = df_iqr[col].quantile(0.25)
    
    # Calculate third quartile (75th percentile)
    Q3 = df_iqr[col].quantile(0.75)
    
    # Compute Interquartile Range (spread of middle 50% data)
    IQR = Q3 - Q1
    
    # Define lower and upper bounds for acceptable values
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # Keep only rows within the acceptable range (remove outliers)
    df_iqr = df_iqr[(df_iqr[col] >= lower) & (df_iqr[col] <= upper)]

# Print shape after removing outliers
print("After IQR:", df_iqr.shape)
```

## Percentile Method

```python
df_percentile = df_outliers.copy()  # Create a copy to preserve original dataset

# Loop through each numerical column to cap extreme values
for col in num_cols:
    
    # Calculate lower bound (1st percentile)
    lower = df_percentile[col].quantile(0.01)
    
    # Calculate upper bound (99th percentile)
    upper = df_percentile[col].quantile(0.99)
    
    # Replace values below lower and above upper with the boundary values
    # This caps outliers instead of removing rows
    df_percentile[col] = np.clip(df_percentile[col], lower, upper)

# Shape remains same because no rows are removed
print("After Percentile Capping:", df_percentile.shape)
```

## Winsorization

```python
df_winsor = df_outliers.copy()  # Create a copy to keep original data unchanged

# Loop through each numerical column
for col in num_cols:
    
    # Apply Winsorization:
    # limits=[0.01, 0.01] means:
    # - lowest 1% values → replaced with 1st percentile value
    # - highest 1% values → replaced with 99th percentile value
    df_winsor[col] = winsorize(df_winsor[col], limits=[0.01, 0.01])

# Shape remains unchanged because values are capped, not removed
print("After Winsorization:", df_winsor.shape)
```

```python
#Before vs After Outlier Treatment

print("Original Shape:", df_outliers.shape)
print("Z-score Shape:", df_zscore.shape)
print("IQR Shape:", df_iqr.shape)
print("Percentile Shape:", df_percentile.shape)
print("Winsor Shape:", df_winsor.shape)
```

## Final Clean Dataset

```python
# ─────────────────────────────────────────────
# Final Clean Dataset
# ─────────────────────────────────────────────

df_final = df.copy()  # Start with a fresh copy

# -------- Missing Value Handling --------

from sklearn.impute import SimpleImputer

# Numerical columns
num_cols = ['age', 'bmi', 'blood_pressure', 'cholesterol', 'glucose']

# Apply median imputation (robust to outliers)
num_imputer = SimpleImputer(strategy='median')
df_final[num_cols] = num_imputer.fit_transform(df_final[num_cols])

# Categorical columns
cat_imputer = SimpleImputer(strategy='most_frequent')
df_final[['region']] = cat_imputer.fit_transform(df_final[['region']])

# Fill gender using mode
df_final['gender'] = df_final['gender'].fillna(df_final['gender'].mode()[0])

# -------- Outlier Handling (IQR Method) --------

# Create a mask to filter all columns together
mask = pd.Series(True, index=df_final.index)

for col in num_cols:
    Q1 = df_final[col].quantile(0.25)
    Q3 = df_final[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # Update mask instead of filtering repeatedly
    mask &= (df_final[col] >= lower) & (df_final[col] <= upper)

# Apply mask once
df_final = df_final[mask]

# -------- Final Check --------

print("Final Shape:", df_final.shape)
print("\nMissing Values:\n", df_final.isnull().sum())
```

# Brief Report

### 1. Most Effective Imputation Strategy

The most effective imputation strategy was **median imputation for numerical features** and **mode (most frequent) imputation for categorical features**. Median is robust to outliers and provides stable central values, making it suitable for features such as age, BMI, cholesterol, and glucose. Mode imputation ensures logical and consistent replacement for categorical variables like gender and region.

Although advanced techniques such as KNN and MICE imputation were applied, they are computationally more complex and were not necessary for this dataset. Simpler methods provided efficient and reliable results.

---

### 2. Best Outlier Handling Method

The **Interquartile Range (IQR) method** was the most effective for handling outliers. It identifies extreme values using the range between the first quartile (Q1) and third quartile (Q3), making it robust to skewed data distributions.

Compared to the Z-score method, IQR does not assume normal distribution and performs better on real-world data. Unlike percentile capping and Winsorization, which only limit extreme values, the IQR method removes unrealistic values entirely, improving data quality.

---

### 3. Improvement in Dataset Usability

Data cleaning significantly improved the usability of the dataset. Missing values were handled appropriately, ensuring completeness across all features. Outliers were treated, reducing noise and improving the reliability of the data.

As a result, the dataset became more consistent, accurate, and suitable for machine learning applications. The cleaned dataset is now well-prepared for tasks such as disease risk prediction.
